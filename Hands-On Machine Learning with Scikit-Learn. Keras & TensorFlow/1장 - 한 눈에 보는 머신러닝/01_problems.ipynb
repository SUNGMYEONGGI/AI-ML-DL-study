# 1장 한눈에 보는 머신러닝

#### 1. 머신러닝을 어떻게 정의할 수 있을까요?
- 머신러닝은 컴퓨터가 데이터를 분석하여 통계적인 규칙을 찾아내고, 이 규칙을 바탕으로 새로운 데이터에 대한 예측을 수행하는 것을 말합니다.
- 머신러닝은 데이터로부터 학습할 수 있는 시스템을 만드는 것입니다. 학습이란 어떤 작업에서 주어진 성능 지표가 더 나아지도록 시스템을 조정하는 것을 말합니다.

#### 2. 머신러닝이 도움을 주는 분야는 무엇이 있을까요?
- 머신러닝은 다양한 분야에서 사용됩니다. 예를 들어, 스팸 메일을 필터링하는 데에도 사용되고, 음성 인식에도 사용되며, 이미지를 분류하는 데에도 사용됩니다.
- 명확한 해결책이 없는 복잡한 문제, 수작업으로 만든 긴 규칙 리스트를 대체하는 경우, 변화하는 환경에 적응하는 시스템을 만드는 경우, 사람에게 통찰을 제공하는 경우에 머신러닝을 사용할 수 있습니다.

#### 3. 레이블된 훈련 세트란 무엇인가요?
- 레이블된 훈련 세트는 각 샘플에 대해 정답이 있는 데이터입니다. 즉, 레이블된 훈련 세트는 입력과 출력이 모두 있는 데이터입니다.
- 레이블된 훈련 세트는 각 샘플에 대해 원하는 정답(레이블)을 담고 있는 훈련 세트입니다.

#### 4. 가장 널리 사용되는 지도 학습 작업은 무엇인가요?
- 가장 널리 사용되는 지도 학습 작업은 분류와 회귀입니다. 분류는 샘플이 어떤 클래스에 속하는지를 예측하는 것이고, 회귀는 샘플의 특성에 따라 연속적인 값을 예측하는 것입니다.
- 가장 일반적인 두 가지 지도 학습 문제는 회귀와 분류입니다.

#### 5. 보편적인 비지도 학습 작업 네 가지는 무엇인가요?
- 보편적인 비지도 학습 작업은 군집, 시각화, 차원 축소, 연관 규칙 학습입니다.

#### 6. 사전 정보가 없는 여러 지형에서 로봇을 걸어가게 하려면 어떤 종류의 머신러닝 알고리즘을 사욜할 수 있을까요?
- 사전 정보가 없는 여러 지형에서 로봇을 걸어가게 하려면 강화 학습을 사용할 수 있습니다.
- 이는 전형적으로 강화 학습이 다루는 유형의 문제입니다. 이 문제를 지도 학습이나 비지도 학습으로 표현하는 것도 가능하지만 일반적이지 않습니다.

#### 7. 고객을 여러 그룹으로 분할하려면 어떤 알고리즘을 사용해야 하나요?
- 고객을 여러 그룹으로 분할하려면 비지도 학습의 군집 알고리즘을 사용해야 합니다.
- 만약 그룹을 어떻게 정의할지 모른다면 비슷한 고객끼리 군집으로 나누기 위해 군집 알고리즘(비지도 하학습)을 사용할 수 있습니다. 그러나 어떤 그룹이 있어야 할지 안다면 분류 알고리즘(지도 학습)에 각 그룹에 대한 샘플을 주입합니다. 그러면 알고리즘이 전체 고객을 이런 그룹으로 분류하게 될 것입니다.

#### 8. 스팸 감자의 문제는 지도 학습과 비지도 학습 중 어떤 문제로 볼 수 있나요?
- 스팸 감자의 문제는 지도 학습 문제입니다. 스팸 감자를 구분하는 데에는 레이블된 훈련 세트가 필요합니다.

#### 9. 온라인 학습 시스템이 무엇인가요?
- 온라인 학습 시스템은 데이터를 순차적으로 한 개씩 또는 미니배치로 주입받아 학습하는 시스템입니다.

#### 10. 외부 메모리 학습이 무엇인가요?
- 외부 메모리 학습은 데이터를 한 번에 하나씩 또는 미니배치로 주입받아 학습하는 시스템입니다. 이 시스템은 훈련 세트가 메모리에 들어갈 만큼 충분히 작지 않을 때 사용합니다.

#### 11. 예측을 하기 위해 유사도 측정에 의존하는 학습 알고리즘은 무엇인가요?
- 예측을 하기 위해 유사도 측정에 의존하는 학습 알고리즘은 사례 기반 학습입니다.
- 사례 기반 학습은 새로운 샘플에 대한 예측을 만들기 위해 훈련 세트에서 가장 비슷한 샘플을 찾습니다. 그리고 이 샘플의 레이블을 예측에 사용합니다.

#### 12. 모델 파라미터와 학습 알고리즘의 하이퍼파라미터 사이에는 어떤 차이가 있나요?
- 모델 파라미터는 모델이 학습하는 동안 결정되는 파라미터이고, 학습 알고리즘의 하이퍼파라미터는 모델이 학습하는 동안 결정되지 않는 파라미터입니다.


#### 13. 모델 기반 알고리즘이 찾는 것은 무엇인가요? 성공을 위해 이 알고리즘이 사용하는 가장 일반적인 전략은 무엇인가요? 예측은 어떻게 만드나요?
- 모델 기반 알고리즘은 모델 파라미터를 찾는 것입니다. 성공을 위해 이 알고리즘이 사용하는 가장 일반적인 전략은 비용 함수를 최소화하는 모델 파라미터를 찾는 것입니다. 예측은 모델 파라미터를 사용해 만듭니다.

#### 14. 머신러닝의 주요 도전 과제는 무엇인가요?
- 머신러닝의 주요 도전 과제는 부족한 데이터, 낮은 데이터 품질, 대표성 없는 데이터, 과대적합, 과소적합, 훈련 시간이 너무 오래 걸리는 문제입니다.

#### 15. 모델이 훈련 데이터에서의 성능은 좋지만 새로운 샘플에서의 일반화 성능이 나쁘다면 어떤 문제가 있는 건가요? 가능한 해결책 세 가지는 무엇인가요?
- 모델이 훈련 데이터에서의 성능은 좋지만 새로운 샘플에서의 일반화 성능이 나쁘다면 과대적합이 발생한 것입니다. 가능한 해결책은 다음과 같습니다.
    - 훈련 데이터를 더 모은다.
    - 모델의 제약을 줄인다.
    - 모델의 파라미터 수를 줄인다.

#### 16. 테스트 세트가 무엇이고 왜 사용해야 하나요?
- 테스트 세트는 모델을 평가하는 데 사용하는 데이터 세트입니다. 테스트 세트를 사용해 모델을 평가하면 모델이 훈련 세트에 과대적합되었는지 확인할 수 있습니다.

#### 17. 검증 세트의 목적은 무엇인가요?
- 검증 세트는 하이퍼파라미터를 조정하는 데 사용합니다. 검증 세트를 사용해 하이퍼파라미터를 조정하면 모델이 훈련 세트에 과대적합되었는지 확인할 수 있습니다.
